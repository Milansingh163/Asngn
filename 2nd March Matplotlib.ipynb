<h3>Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.</h3>
<p>Matplotlib is a popular data visualization library in Python that is used to create high-quality static, animated, and interactive plots.

Matplotlib is used for a wide range of data visualization tasks, including exploratory data analysis, scientific and engineering visualizations, and creating high-quality publication-ready graphics. It provides a comprehensive set of tools for creating a wide range of plot types and styles, making it a versatile library for any data visualization task.

Here are five common plot types that can be created using the Pyplot module of Matplotlib:

1. Line plots: A line plot is a basic plot that displays data as a series of points connected by straight lines.

2. Scatter plots: A scatter plot is a plot that displays data as a collection of points, with each point representing the values of two variables.

3. Bar plots: A bar plot is a plot that displays data as a series of bars, with the height of each bar representing the value of a variable.

4. Histograms: A histogram is a plot that displays the distribution of a single variable by dividing the data into a set of bins and counting the number of data points in each bin.

5. Heatmaps: A heatmap is a plot that displays data as a matrix of colored squares, with the color of each square representing the value of a variable at a particular point in time or space.</p>

<h3>Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.</h3>
<p>
import numpy as np <br>
np.random.seed(3)<br>
x = 3 + np.random.normal(0, 2, 50)<br>
y = 3 + np.random.normal(0, 2, len(x))<br>
Note: Also add title, xlabel, and ylabel to the plot. </p>

<h6>Answer : </h6><br>
<p>
A scatter plot is a type of plot in which the values of two variables are plotted as a set of points on a two-dimensional coordinate system, with one variable on the x-axis and the other variable on the y-axis. Each point on the plot represents the values of both variables for a single observation or data point.

Scatter plots are useful for visualizing the relationship between two variables and identifying any patterns or trends in the data. For example, a scatter plot can be used to determine whether there is a positive or negative correlation between two variables, or whether there is no correlation at all. Scatter plots can also be used to identify outliers, which are data points that fall far from the main cluster of points on the plot.

Scatter plots can be customized in various ways, including changing the color, size, and shape of the points, adding labels and titles, and adjusting the axes and scales. They are commonly used in data analysis and visualization, as well as in scientific research and engineering applications.

</p>

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x, y)
plt.title('Simple scatter plot')
plt.xlabel('Value for x')
plt.ylabel('Value for y')
plt.show()
```

<h3>Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:</h3>
<p>import numpy as np <br>
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])<br>
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])<br>
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])<br>
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])<br></p>
<h6>Answer : </h6>
<p>The subplot() function is used to create multiple plots within a single figure. It takes three arguments: the number of rows, the number of columns, and the index of the current subplot.

The index of the subplot is counted row by row from left to right and from top to bottom. For example, if the subplot has 2 rows and 3 columns, the first subplot is in the top-left corner, the second subplot is in the top-middle, and so on.</p>

```python
# Data for line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure with four subplots
fig, axs = plt.subplots(2, 2)

# Plot line 1
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')

# Plot line 2
axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')

# Plot line 3
axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')

# Plot line 4
axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')

# Add a title for the whole figure
fig.suptitle('Four Line Plots')

# Show the plots
plt.show()
```

<h3>Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.</h3>
import numpy as np <br>
company = np.array(["Apple", "Microsoft", "Google", "AMD"]) <br>
profit = np.array([3000, 8000, 1000, 10000]) <br>
<h6>Answer</h6>
<p>
A bar plot is a type of plot in which data is represented using rectangular bars, where the height or length of the bars corresponds to the values of the data. Bar plots are commonly used to compare the values of different categories or groups.

Bar plots can be used to visualize various types of data, such as categorical data, numerical data, and time-series data. They are especially useful for displaying data that can be divided into discrete categories or groups.
</p>

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
plt.title('Barplot')
plt.xlabel('X axis')
plt.ylabel('y axis')
plt.show()
```

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.barh(company,profit)
plt.title('Barplot')
plt.xlabel('X axis')
plt.ylabel('y axis')
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
</h3>

box1 = np.random.normal(100, 10, 200)<br>
box2 = np.random.normal(90, 20, 200)<br>

A box plot is a type of plot that displays the distribution of a dataset. It is also known as a box-and-whisker plot, as it consists of a box that represents the interquartile range (IQR) of the data, along with whiskers that extend to the minimum and maximum values of the data that are not outliers. Outliers are typically represented by dots or other symbols outside of the whiskers.

Box plots are useful for visually summarizing the distribution of a dataset, especially when comparing multiple datasets or groups. They can also be used to detect outliers in the data and to compare the spread and central tendency of the data between different groups.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.boxplot([box1,box2])
plt.title('Box Plot of Two Datasets')
plt.xticks([1, 2], ['Box 1', 'Box 2'])
plt.ylabel('Values')
plt.show()
```
