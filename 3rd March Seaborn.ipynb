<h3>Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.<br>
</h3>
<h6>Answer : </h6>
<p>
1. Scatter Plot: A scatter plot shows the relationship between two numerical variables. It can be used to identify patterns and relationships between variables.

2. Line Plot: A line plot displays a relationship between two variables with a continuous line. It is useful for visualizing trends over time or a continuous range.

3. Histogram: A histogram shows the distribution of a single numerical variable. It can be used to identify the frequency and distribution of values in a dataset.

4. Bar Plot: A bar plot compares categorical variables with numerical data. It can be used to compare data across categories.

5. Heatmap: A heatmap shows the relationship between two categorical variables or a categorical and numerical variable. It is useful for identifying patterns and relationships between variables.</p>

<h3>Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.<br>
</h3>
Note: timepoint, signal, event, and region are columns in the fmri dataset.<br>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fmri =  sns.load_dataset('fmri')
fmri.head()
```

```python
sns.lineplot(data=fmri,x='timepoint',y='signal',hue='region',style='event')
plt.show()
```

<h3>Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
</h3>
Note: pclass, age, and fare are columns in the titanic dataset.

```python
titanic =  sns.load_dataset('titanic')
titanic.head()
```

```python
sns.boxplot(data=titanic,x = 'pclass',y='age')
plt.show()
```

```python
sns.boxplot(data=titanic,x = 'pclass',y='fare')
plt.show()
```

<h3>Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.
</h3>

```python
diamonds =  sns.load_dataset('diamonds')
diamonds.head()
```

```python
plt.figure(figsize=(8,7))
sns.histplot(data=diamonds,x='price', hue='cut',multiple="stack")
plt.show()
```

<h3>Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.</h3>

```python
iris =  sns.load_dataset('iris')
iris.head()
```

```python
sns.pairplot(data=iris,hue='species')
plt.show()
```

<h3>Que 6: Use the "flights" dataset from seaborn to plot a heatmap.</h3>

```python
flights =  sns.load_dataset('flights')
flights.head()
```

```python
plt.figure(figsize=(10,7))
heatmap_data = flights.pivot('year','month','passengers')
sns.heatmap(heatmap_data,annot=True, fmt=".1f")
plt.show()
```
