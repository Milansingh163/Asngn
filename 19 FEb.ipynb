<h3>Q1. What is Flask Framework? What are the advantages of Flask Framework? </h3>

<p>
Flask is a lightweight web framework written in Python. It is designed to be easy to use and flexible, and it allows developers to quickly create web applications. Flask is a "micro" framework, which means it provides only the essential features necessary for building web applications, leaving out many of the extra features that come with larger frameworks.

Some advantages of Flask framework are:

1. Lightweight: Flask is a lightweight framework and does not require any particular tools or libraries to run. This makes it easy to get started and allows developers to build simple, efficient web applications quickly.

2. Flexible: Flask is a flexible framework that allows developers to customize and extend it as needed. It does not dictate a particular project structure, allowing developers to organize their code as they see fit.

3. Easy to learn: Flask has a relatively small learning curve, especially for developers who are already familiar with Python. Its simple syntax and intuitive design make it easy for developers to get started and build web applications quickly.

4. Modular: Flask is designed to be modular, allowing developers to use only the components they need. This can help to reduce the size and complexity of web applications, making them faster and easier to maintain.

5. Extensible: Flask can be extended with a wide range of third-party libraries and plugins, making it easy to add features such as authentication, database integration, and more.

Overall, Flask is an excellent choice for developers who want a lightweight, flexible, and easy-to-use framework for building web applications in Python.
</p>

<h3>Q3. What is App routing in Flask? Why do we use app routes?</h3>

<p>In Flask, app routing is the process of mapping URLs to functions that handle requests to those URLs. App routes are used to define these mappings and specify the functions that should be called to handle each URL.</p>
<p>
App routes are important in Flask because they allow developers to create dynamic web applications that can respond to different URLs and user inputs. By mapping URLs to specific functions, Flask can route requests to the appropriate handlers and generate appropriate responses.
</p>
